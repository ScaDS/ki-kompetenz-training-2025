```python
import pandas as pd
import umap
import stackview
import numpy as np
import yaml
```

```python
from wordcloud import STOPWORDS
STOPWORDS.update(("im um die und zu für das von zur ist sowie bie den der durch oder bei zudem person wie des sie er eine mit anderen andere über darüber hinaus diese ein ich verantwortlich koordination mitarbeiter auch einer Doktorarbeit auf könnte sein").split(" "))
```

```python
# Load the DataFrame from the YAML file
with open("scientists_dataset_curated_embedded.yml", 'r') as file:
    data_dict = yaml.safe_load(file)
df = pd.DataFrame(data_dict)
df.head()
```

```python
stackview.scatterplot(df, column_x="UMAP0", column_y="UMAP1")
```

```python
stackview.wordcloudplot(df, column_text="topic", column_x="UMAP0", column_y="UMAP1")
```

```python
stackview.wordcloudplot(df, column_text="research_field", column_x="UMAP0", column_y="UMAP1")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a larger figure size
plt.figure(figsize=(12, 8))

# Create the scatterplot
sns.scatterplot(data=df, x='UMAP0', y='UMAP1', hue='research_field')

# Adjust the font size of axis labels
plt.xlabel('UMAP0', fontsize=14)
plt.ylabel('UMAP1', fontsize=14)

# Adjust the legend
plt.legend(fontsize='small', bbox_to_anchor=(0, 1), loc='upper left')

plt.savefig("faculties.png")

# Show the plot
plt.show()
```
